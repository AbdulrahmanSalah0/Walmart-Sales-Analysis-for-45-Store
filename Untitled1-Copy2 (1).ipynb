```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_csv("Data.csv")
ploted_data = pd.read_csv("Data.csv")
ploted_data = ploted_data.drop(['Store'], axis=1)
```

```python
data['year']=pd.to_datetime(data['Date'],format="%d-%m-%Y").dt.year
data['month']=pd.to_datetime(data['Date'],format="%d-%m-%Y").dt.month
data['day']=pd.to_datetime(data['Date'],format="%d-%m-%Y").dt.day
data['quarter']=pd.to_datetime(data['Date'],format="%d-%m-%Y").dt.quarter
data['semester']=np.where(data['quarter'].isin([1,2]),1,2)
data 
```

```python
df = data.groupby('Store')['Weekly_Sales'].sum().to_frame().reset_index()
Max_Sales = df[(df['Weekly_Sales']==max(df['Weekly_Sales']))]
Max_Sales
```

```python
df = data.groupby('Store')['Weekly_Sales'].std().to_frame().reset_index()
standard_deviation = df[(df['Weekly_Sales']==max(df['Weekly_Sales']))]
standard_deviation
```

```python
non_Holidays = data[(data['Holiday_Flag']==0)]
Mean = non_Holidays.mean()
Mean_NH = Mean['Weekly_Sales']
Mean_NH
```

```python
Holidays = data[(data['Holiday_Flag']==1)]
Holidays_Bigger_Than_Mean =Holidays[(Holidays['Weekly_Sales']>Mean_NH)]
Holidays_Bigger_Than_Mean
```

```python
profit_Of_each_Month =data.groupby(['year','month'])['Weekly_Sales'].sum().to_frame().reset_index()
profit_Of_each_Month
```

```python
profit_Of_each_semester =data.groupby(['year','semester'])['Weekly_Sales'].sum().to_frame().reset_index()
profit_Of_each_semester
```

```python
Weekly_sales_sum =data.groupby('Store')["Weekly_Sales"].sum().reset_index()
plt.bar(Weekly_sales_sum["Store"],Weekly_sales_sum["Weekly_Sales"])
plt.xlabel("stores")
plt.ylabel("weekly sales")
plt.title("weekly sales comparison")
plt.show()
```

```python
plt.boxplot(data["Unemployment"])
plt.title("Unemployment")
plt.show()
```

```python
plt.boxplot(data["CPI"])
plt.title("CPI ")

plt.show()
```

```python
year_view_of_sales=data.groupby(["year"])["Weekly_Sales"].sum().reset_index()
year_view_of_sales
```

```python
plt.pie(year_view_of_sales["Weekly_Sales"],labels=year_view_of_sales["year"])
plt.title("comparison between the sales in years")
plt.show()
```

```python
sales_and_fuel=data.groupby("Store").agg({
 'Weekly_Sales': lambda sales : sales.sum(),
 'Fuel_Price' : lambda fuel : fuel.mean()
}).reset_index()
plt.scatter(sales_and_fuel["Store"],sales_and_fuel["Weekly_Sales"],s=sales_and_fuel["Fuel_Price"],alpha=1)
plt.show()
# fuel price has no effect in the weekly sales
```

```python
sales_and_temp=data.groupby("Store").agg({
 'Weekly_Sales': lambda sales : sales.sum(),
 'Temperature' : lambda temp : temp.mean()
}).reset_index()
plt.scatter(sales_and_temp["Store"],sales_and_temp["Weekly_Sales"],s=sales_and_temp["Temperature"],alpha=1)
plt.show()
# temperature has no effect in the weekly sales
```
